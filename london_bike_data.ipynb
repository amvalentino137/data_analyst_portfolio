```python
# Importing packages
import calendar
import pandas as pd
```

```python
# Notes from dataset
'''"timestamp" - timestamp field for grouping the data
"cnt" - the count of a new bike shares
"t1" - real temperature in C
"t2" - temperature in C "feels like"
"hum" - humidity in percentage
"wind_speed" - wind speed in km/h
"weather_code" - category of the weather
"is_holiday" - boolean field - 1 holiday / 0 non holiday
"is_weekend" - boolean field - 1 if the day is weekend
"season" - category field meteorological seasons: 0-spring ; 1-summer; 2-fall; 3-winter.

"weather_code" category description:
1 = Clear ; mostly clear but have some values with haze/fog/patches of fog/ fog in vicinity 
2 = scattered clouds / few clouds 
3 = Broken clouds 
4 = Cloudy 
7 = Rain/ light Rain shower/ Light rain 1
0 = rain with thunderstorm 
26 = snowfall 
94 = Freezing Fog'''
```

```python
# reading in data
df = pd.read_csv(r'/Users/alecvalentino/Desktop/Data Analytics Bootcamp/london_merged.csv')
df.info()
```

```python
# Exploring data
df.shape
```

```python
# Changing column names to more clear headers
new_col_dict = {
    'timestamp':'time',
    'cnt':'count',
    't1':'temp_real_C',
    't2': 'temp_feels_like_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}
#renaming the columns to the new column names
df.rename(new_col_dict, axis = 1, inplace = True)
df
```

```python
# Changing humidity value to humidity percentage
df.humidity_percent = df.humidity_percent / 100
df
```

```python
# creating a dictionary to define seasons
seasons_dict = {
    '0.0':'spring',
    '1.0': 'summer',
    '2.0': 'autumn',
    '3.0':'winter'
}
# Creating a weather dictionary 
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall',
    '94.0':'Freezing Fog'
}
# Changing the seasons column data type to string
df.season = df.season.apply(str)
# replacing numeric values with string values
df.season = df.season.map(seasons_dict)

#changing the weather column data type to string
df.weather = df.weather.apply(str)
# replacing numeric values with string values
df.weather = df.weather.map(weather_dict)
```

```python
df.head()
```

```python

# Splitting Time Stamp Column into Date and Time
df[['Date', 'Time']] = df['time'].str.split(' ', expand = True)
df
```

```python
# Splitting Date into Year, Month, and Day
df[['Year','Month','Day']] = df['Date'].str.split('-', expand = True)
df
```
