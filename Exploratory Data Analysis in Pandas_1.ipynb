```python
# group together columns and look at the data
df.groupby('Continent').mean().sort_values(by= '2022 Population', ascending= False)
```

```python
# we want to visualize our columns by continent, but if we just use our dataset like normal, it won't look right
# if we tranpose our data we can get a better result
df2 = df.groupby('Continent')[['2022 Population',
       '2020 Population', '2015 Population', '2010 Population',
       '2000 Population', '1990 Population', '1980 Population',
       '1970 Population']].mean().sort_values(by= '2022 Population', ascending= False)
df3= df2.transpose()
df2.plot()
df3.plot()
```

```python
# if we need a quick grab of the columns in a list, we can use df.columns
df.columns
```

```python
# this allows us to insert this list here
df2 = df.groupby('Continent')[['1970 Population','1980 Population','1990 Population','2000 Population','2010 Population','2015 Population','2020 Population','2022 Population']].mean().sort_values(by= '2022 Population', ascending= False)
df2
```

```python
# now with our optimized code, we can run our previous tranpose
df3= df2.transpose()
df3
```
